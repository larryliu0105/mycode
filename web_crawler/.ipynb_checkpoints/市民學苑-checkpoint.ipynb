```python
import requests
res = requests.get("http://163.32.98.3/KHHCVS/Main/ClassList.aspx?new=1")
res.close()
res.encoding = res.apparent_encoding
```

```python
from bs4 import BeautifulSoup
soup = BeautifulSoup(res.text, "lxml")
print(soup.prettify())
```

```python
table = soup.select('#DataRow')
print(table)
```

```python
text= ""
for item in table:
    href = "null"
    for idx, cell in enumerate(item.select('td')):
        if idx == 13 and cell.text.strip()=='':
            text += 'empty'+','
        else:
            text += cell.text.strip()+','
        if idx == 4 and cell.text.strip() != '':
            href = cell.find('a').get('href')
    text += href
    text += '\n'
print(text)
```

```python
title=''
header = soup.select('#HeaderRow')
for i in header[0].select('td'):
    title += i.text+','
title += '師資網址'
print (title)
```

```python
with open('output3.csv','w', encoding ='utf8') as outputfile:
    outputfile.write(title+'\n')
    outputfile.write(text)
```
