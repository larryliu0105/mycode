```python
import requests
res = requests.get("http://www.kmseh.gov.tw/07_class/online.php")
res.close()
res.encoding = res.apparent_encoding
```

```python
from bs4 import BeautifulSoup
soup = BeautifulSoup(res.text, "lxml")
print(soup.prettify())
```

```python
table = soup.find('table',{"summary":"硬體設備"})
text= ""
href = "詳細資料"
for idx1, row in enumerate(table.findAll("tr")):
    aaa =  row.findAll("th")
    cells = row.findAll("td")
    cells = aaa+cells
    if len(cells) == 10:
        for idx2, i in enumerate(cells):
            text += i.text.strip()+','
            if idx2 ==1 and idx1!=0:
                href = i.find('a').get('href')
        text += href+"\n"
print (text)

        
```

```python
with open('output2.csv','w', encoding ='utf8') as outputfile:
    outputfile.write(text)
```
