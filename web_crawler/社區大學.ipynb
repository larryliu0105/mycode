```python
import requests
res = requests.get("http://kh1cu.blogspot.tw/p/105.html")
res.close()
res.encoding = res.apparent_encoding
```

```python
from bs4 import BeautifulSoup
soup = BeautifulSoup(res.text, "lxml")
print(soup.prettify())
```

```python
table = soup.select('#Blog1')
text= ""
num = -1
for idx, i in enumerate(table[0].select('td')):
    content = i.text.replace('\n','').strip()
    if content.startswith("星期"):
        text += content+'\n'
        num += 1
    elif (idx-num)%6 ==0:
        text += content+'\n'
    else:
        text += content+','
print (text)
```

```python
with open('output2.csv','w', encoding ='utf8') as outputfile:
    outputfile.write(text)
```
